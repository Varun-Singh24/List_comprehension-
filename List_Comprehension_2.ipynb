```python
ls = [1,2, 3, 4, 5]
```

To find out the square of the above list We use list comprehension

```python
# Using the loop
for i in ls :
  print(i**2, end=" ")
```

```python
# using the list Comprehension
sq_wid_comprehension = [i**2 for i in ls]
```

```python
sq_wid_comprehension
```
